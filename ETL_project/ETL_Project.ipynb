```python
# Import dependencies
import pandas as pd
from sqlalchemy import create_engine
import datetime as dt
import numpy as np
```

```python
avocado_file = "avocado.csv"
```

```python
avocado_df = pd.read_csv(avocado_file)
avocado_df.head()
```

```python
new_avocado_df = avocado_df[['Date', 'AveragePrice', 'type', "region"]].copy()
new_avocado_df.head()
```

```python
con_avocado_df= new_avocado_df.loc[new_avocado_df['type']=="conventional"]
con_avocado_df.head()
```

```python
con_avocado_df["Date"] = pd.to_datetime(con_avocado_df["Date"])
```

```python
con_avocado_df['Date'] = con_avocado_df.Date + pd.Timedelta(days=1)
con_avocado_df.head()
```

```python
ave_avocado_df = con_avocado_df.groupby("Date")["AveragePrice"].mean()
ave_df = pd.DataFrame(ave_avocado_df)
ave_df.head()
```

```python
ave_df.reset_index(inplace=True)
ave_df.head()
```

```python
ave_df = ave_df.rename(columns={"Date": "date",
                               "AveragePrice":"averageprice"})
ave_df.head()
```

```python
CMG_file = "CMG_data.csv"
```

```python
CMG_df = pd.read_csv(CMG_file)
CMG_df.head()
```

```python
CMG_df["date"] = pd.to_datetime(CMG_df["date"])
```
